```python
# AS THERE ARE STRING AND FLOAT DATA TYPE COLUMNS, SEPERATING THEM FOR BETTER DATA ANALYSIS
```

```python
# BAR CHART ON LOW POPULATED COUNTRIES IN 2023
```

```python
low_pop_countries = world_pop_data_upd.loc[:, ('Country Name', '2023')].set_index('Country Name')
low_pop_countries = low_pop_countries.sort_values(by='2023', ascending=False).tail(5)

low_pop_countries
```

```python
def plot_low_pop_countries_2023(data_2023):
    """
    Plots a bar chart of the world's low-population countries for the year 2023.

    Functionality
    -------------
    Creates a bar plot showing each country's population in billions for 2023.
    The x-axis represents country names, and the y-axis represents population.
    """
    plt.figure(figsize=(6,5))
    plt.bar(data_2023.index, data_2023['2023'], color='green', edgecolor='black')
    plt.title("WORLD'S LOW POPULATED COUNTRIES IN 2023")
    plt.xlabel('Country Name')
    plt.xticks(rotation=50, fontsize=6)
    plt.yticks(fontsize=6)
    plt.ylabel('Population (in billions)')
    plt.show()
```

```python
plot_low_pop_countries_2023(low_pop_countries)
```

```python
# DRWAING LINE PLOT FOR FEW COUNTRIES SHOWING POPULATION ACROSS YEARS
```

```python
countries_list = [
    'Myanmar', 'India', 'Finland', 'United States', 'Africa Western and Central'
]

countries_data = (
    world_pop_data_upd
    .set_index('Country Name')
    .loc[countries_list, '2014':]
)

countries_data
```

```python
# TAKING COUNTRIES ON X-AXIS AND YEARS ON Y-AXIS
```

```python
Transpose_data = countries_data.T

Transpose_data
```

```python
# FUNCTION TO DRAW LINE PLOT FOR ABOVE DATA
```

```python
def pop_growth_comparison(data):
    """
    Plots a line chart to compare population growth over the last ten years for a list of countries.

    Functionality
    -------------
    For each country in countries_list, this function plots a line representing its population growth
    over the last ten years. Each line is labeled with the respective country name, and the chart includes 
    gridlines, a legend, and formatted x-axis labels for clarity. 
    """
    
    plt.figure(figsize=(6,5))
    for country in countries_list:
        plt.plot(data.index, data[country], label=country)
    plt.title('Population Trends of Selected Countries In Last Ten Years')
    plt.xlabel('Year')
    plt.ylabel('Population (in billions)')
    plt.xticks(rotation=30)
    plt.legend(loc=10)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

pop_growth_comparison(Transpose_data)
```

```python
# CORRELATING SAMPLE DATA AND DRAWING HEATMAP
```

```python
correlation_data = numerical_data.loc[:, '2014':].corr()

correlation_data
```

```python
def heatmap(corr_data):
    """
    Plots a heatmap to visualize the correlation matrix of population data over the years from 2014 to 2023.

    Functionality
    -------------
    Creates a heatmap of the correlation matrix using a cool-warm color palette. The heatmap
    displays correlations without annotations on individual cells, and includes a title.
    """
    plt.figure(figsize=(6,5))
    sns.heatmap(corr_data, annot=False, cmap='coolwarm', linewidth=0.5)
    plt.title('Population Correlation Matrix (2014–2023)')
    plt.xticks(fontsize=8)
    plt.yticks(fontsize=8)
    plt.show()

heatmap(correlation_data)
```

```python
import pandas as pd
from scipy.stats import skew, kurtosis

# Dropping NA values and selecting numeric columns
stats_data = world_pop_data_upd.dropna(axis=0, how='any').set_index('Country Name')

# Selecting only numeric data
numeric_data = stats_data.loc[:, '2014':].apply(pd.to_numeric, errors='coerce')

# Calculating skewness and kurtosis
skewness = numeric_data.apply(skew)
kurtosis_values = numeric_data.apply(kurtosis)

# Results kept in a DataFrame
skewness_kurtosis_df = pd.DataFrame({
    'Year': numeric_data.columns,
    'Skewness': skewness,
    'Kurtosis': kurtosis_values
})

print(skewness_kurtosis_df)
```
